```python
# we need to do one hot encoding in sex , smoker , region
# for that we need to separate the column
sex = df["sex"]
region = df["region"]
smoker = df["smoker"]

print(sex.shape)
print(region.shape)
print(smoker.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder

sex_encoder = OneHotEncoder(drop="first", sparse_output=False)
region_encoder = OneHotEncoder(drop="first", sparse_output=False)
smoker_encoder = OneHotEncoder(drop="first", sparse_output=False)


sex_encoded = sex_encoder.fit_transform(sex.to_frame()) # because sklear needs 2d data there fore it sex.to_frame
region_encoded = region_encoder.fit_transform(region.to_frame())
smoker_encoded = smoker_encoder.fit_transform(smoker.to_frame())


print(sex_encoded)
print(region_encoded)
print(smoker_encoded)
```

```python

print(sex.shape)
sex_2d=(sex.to_frame())
print(sex_2d.shape)
```

```python
df.head()
```

```python
sex_df = pd.DataFrame(sex_encoded,columns=["sex"])
region_df = pd.DataFrame(region_encoded,columns=["region_A","region_B","region_C"])
smoker_df = pd.DataFrame(smoker_encoded,columns=["smoker"])

print(sex_df)
print(region_df)
print(smoker_df)
```

```python
df["sex_encoded"]=sex_df
df["region_A"]=region_df["region_A"]
df["region_B"]=region_df["region_B"]
df["region_C"]=region_df["region_C"]
df["smoker_encoded"]=smoker_df
df.head(10)
```

```python
df.drop(columns=["sex","region","smoker"],inplace=True)
```

```python
Y = df["charges"]
X = df.drop(columns="charges")
# separaet features and target variable
print(Y)
print(X)
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = ["age" , "children" ,"bmi"]

X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])


```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score , mean_squared_error
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)

y_pred = model.predict(X_test)
```

```python
print("R² score:", r2_score(Y_test, y_pred))
mse = mean_squared_error(Y_test, y_pred)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
```
